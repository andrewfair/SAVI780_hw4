```python
dfXwalk = pd.read_csv('../data/national_county.txt', header=None)
```

```python
dfXwalk.head()
```

```python
dfXwalk.columns = ['STATE', 'STATEFP', 'COUNTYFP', 'COUNTYNAME', 'CLASSFP']
```

```python
dfXwalk.head()
```

```python
dfXwalkUniques = dfXwalk[['STATE', 'STATEFP']].drop_duplicates()
```

```python
dfXwalkUniques['STATEFP'] = dfXwalkUniques['STATEFP'].apply('{:0>2}'.format)
```

```python
dfXwalkUniques
```

```python
d={}
for tag,group in dfXwalk.set_index('STATEFP').groupby(level='STATEFP'):
    d[str(tag).zfill(2)]=group.to_dict(orient='records')
print(json.dumps(d,indent=1))
```

```python
d['01'][50]
```

```python
with open('../data/xwalk.json', 'w') as outfile:
    json.dump(d, outfile)
```
